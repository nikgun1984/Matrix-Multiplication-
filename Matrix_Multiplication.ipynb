```python
#PART3
import numpy as np

def split(m):
    #Partitioning of A and B into parts by dividing it in halfs and then using recurssion 
    length = len(m)
    if(length % 2 is not 0):
        stack = []
        for x in range(length + 1):
            stack.append(float(0))
        length += 1
        m = np.insert(m, len(m), values=0, axis=1)
        m = np.vstack([m, stack])
    d = (length // 2)
    m = m.reshape(length, length)
    partition = [m[:d, :d], m[d:, :d], m[:d, d:], m[d:, d:]]
    return partition
def strassen(a, b):
    global operations
    operations+=1
    q = len(a)
    if q == 1:  # base case: 1x1 matrix
        #print(len(a))
        #print(len(b))
        return a*b
    a11, a12, a21, a22 = split(a)#PARTITIONING A Matrix
    #print(a11)
    #print(a12)
    #print(a21)
    #print(a22)
    b11, b12, b21, b22 = split(b)#PARTITIONING B Matrix
    #RECURSIVE WAY TO GET OUR VALUES
    m1 = strassen((a11 + a22), (b11 + b22)) # p1 = (a11 + a22) * (b11 + b22)
    m2 = strassen(a21 + a22, b11)       # p2 = (a21 + a22) * b11
    m3 = strassen(a11, b12 - b22)      # p3 = a11 * (b12 - b22)
    m4 = strassen(a22, b21 - b11)        # p4 = a22 * (b21 - b11)
    m5 = strassen(a11 + a12, b22)        # p5 = (a11 + a12) * b22
    m6 = strassen(a21 - a11, b11 + b12)  # p6 = (a21 - a11) * (b11 + b12)
    m7 = strassen(a12 - a22, b21 + b22)  # p7 = (a12 - a22) * (b21 + b22)
    c11 = m1 + m4 - m5 + m7  # c11 = p1 + p4 - p5 + p7
    c12 = m3 + m5            # c12 = p3 + p5
    c21 = m2 + m4            # c21 = p2 + p4
    c22 = m1 + m3 - m2 + m6  # c22 = p1 + p3 - p2 + p6
    c = np.vstack((np.hstack((c11, c12)), np.hstack((c21, c22)))) 
    return c

operations = 0
a = np.random.uniform(-2, 2, size=(1024, 1024))
print(a.shape)
b = np.random.uniform(-2, 2, size=(1024, 1024))
print(b.shape)
#Strassen Algorithm Test
print(np.array(strassen(a, b)))
print(operations)
```
